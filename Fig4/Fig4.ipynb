## Figure 4: Computation Time Using Different Methods
Data are run using Jupyter Notebooks `SSA_timing.ipynb`, `FSP_timing.ipynb`, `Renewal_timing.ipynb` and `Main_timing.ipynb` and saved to `.npy` files. This Jupyter Notebook creates Figure 4 using the data.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# load results from `.npy` files
FSP_times = np.load('FSP_times.npy')
FSP_times = FSP_times.tolist()
SSA_times = np.load('SSA_times.npy')
SSA_times = SSA_times.tolist()
Renewal_times = np.load('Renewal_times.npy')
Renewal_times = Renewal_times.tolist()
Main_times = np.load('Main_times.npy')
Main_times = Main_times.tolist()
```

```python
print(f'Finite State Projection algorithm for models up to order {len(FSP_times)}\n'
      f'Stochastic Simulation algorithm for models up to order {len(SSA_times)}\n'
      f'Renewal Theory methods for models up to order {len(Renewal_times)}\n'
      f'Our Result for models up to order {len(Main_times)}')
```

```python
print(FSP_times)
print(SSA_times)
print(Renewal_times)
print(Main_times)
```

```python
fig, axs = plt.subplots(2,2, figsize=(30,20))
ax1=axs[0,0]
ax2=axs[0,1]
ax3=axs[1,0]
ax4=axs[1,1]

dim1=list(range(1,11))
dim2=list(range(1,21))
dim3=list(range(1,51))
dim4=list(range(1,201))

# plt.style.use('ggplot')
plt.style.use('default')
plt.rcParams['axes.linewidth'] = 3

sns.lineplot(x=dim1, y=FSP_times[0:10], ax=ax1, label='FSP', color='#DAA520',
             linestyle='--', marker='D', markerfacecolor='#32CD32',
             markersize=20, linewidth=6)
sns.lineplot(x=dim1, y=SSA_times[0:10], ax=ax1, label='SSA', color='#008FFF',
             linestyle='-.', marker='s', markerfacecolor='#6A5ACD',
             markersize=20, linewidth=6)
sns.lineplot(x=dim1, y=Renewal_times[0:10], ax=ax1, label='Renewal Theory', color='red',
             linestyle=':', marker='o', markerfacecolor='blue',
             markersize=20,linewidth=6)
sns.lineplot(x=dim1, y=Main_times[0:10], ax=ax1, label='Our Result', markerfacecolor='#FF6347',
             linestyle='-', marker='v', color='#20B2AA',
             markersize=20, linewidth=6)

# ax1.set_xlabel(f'Matrix Dimension', fontsize=40, fontweight='bold',
#                labelpad=0, fontname='Times New Roman', color='green')
ax1.set_ylabel(f'Computation Time', fontsize=40, fontweight='bold',
               labelpad=10, fontname='Times New Roman', color='green')

ax1.tick_params(axis='y', direction='out', width=3, length=10)
ax1.tick_params(axis='x', direction='out', width=3, length=10)

float_list1 = list(np.arange(0, 50, 10))
tick_labels1 = [f"{value:.1f}s" for value in float_list1]
ax1.set_yticks(float_list1)
ax1.set_yticklabels(tick_labels1, rotation=0, fontsize=30, color='blue',
                    fontname='Times New Roman', ha='right')

ax1.set_xticks(dim1)
ax1.set_xticklabels(dim1, rotation=0, fontsize=40, color='blue',
                    fontname='Times New Roman', ha='right', fontweight='bold')

lines1, labels1 = ax1.get_legend_handles_labels()
legend = ax1.legend(lines1, labels1, loc='upper left',
                    frameon=True, shadow=True, markerscale=1,
                    fancybox=True, prop={'family': 'Times New Roman', 'size': 30})
legend.get_frame().set_color('#E6E6FA')
legend.get_frame().set_alpha(0.8)
ax1.grid(True, which='both', linestyle='-', linewidth=2, alpha=0.3) 

# ********************************************************************************************
sns.lineplot(x=dim2, y=FSP_times[0:20], ax=ax2, label='FSP', color='#DAA520',
             linestyle='--', marker='D', markerfacecolor='#32CD32',
             markersize=20, linewidth=6)
sns.lineplot(x=dim2, y=Renewal_times[0:20], ax=ax2, label='Renewal Theory', color='red',
             linestyle=':', marker='o', markerfacecolor='blue',
             markersize=20,linewidth=6)
sns.lineplot(x=dim2, y=Main_times[0:20], ax=ax2, label='Our Result', markerfacecolor='#FF6347',
             linestyle='-', marker='v', color='#20B2AA',
             markersize=20, linewidth=6)

# ax2.set_xlabel(f'Matrix Dimension', fontsize=40, fontweight='bold',
#                labelpad=0, fontname='Times New Roman', color='green')
# ax2.set_ylabel(f'Computation Time', fontsize=40, fontweight='bold',
#                labelpad=10, fontname='Times New Roman', color='green')

ax2.tick_params(axis='y', direction='out', width=3, length=10)
ax2.tick_params(axis='x', direction='out', width=3, length=10)

float_list2 = list(np.arange(0, 10, 1))  
tick_labels2 = [f"{value:.1f}s" for value in float_list2]
ax2.set_yticks(float_list2)
ax2.set_yticklabels(tick_labels2, rotation=0, fontsize=30, color='blue',
                    fontname='Times New Roman', ha='right')

ax2.set_xticks([1]+list(range(4,24,4)))
ax2.set_xticklabels([1]+list(range(4,24,4)), rotation=0, fontsize=40, color='blue',
                    fontname='Times New Roman', ha='right', fontweight='bold')

lines2, labels2 = ax2.get_legend_handles_labels()
legend2 = ax2.legend(lines2, labels2, loc='upper left',
                    frameon=True, shadow=True, markerscale=1,
                    fancybox=True, prop={'family': 'Times New Roman', 'size': 30})
legend2.get_frame().set_color('#E6E6FA')
legend2.get_frame().set_alpha(0.8)
ax2.grid(True, which='both', linestyle='-', linewidth=2, alpha=0.3)
# ********************************************************************************************
sns.lineplot(x=dim3, y=Renewal_times[0:50], ax=ax3, label='Renewal Theory', color='red',
             linestyle=':', marker='o', markerfacecolor='blue',
             markersize=10,linewidth=6)
sns.lineplot(x=dim3, y=Main_times[0:50], ax=ax3, label='Our Result', markerfacecolor='#FF6347',
             linestyle='-', marker='v', color='#20B2AA',
             markersize=10, linewidth=6)

ax3.set_xlabel(f'Matrix Dimension', fontsize=40, fontweight='bold',
               labelpad=0, fontname='Times New Roman', color='green')
ax3.set_ylabel(f'Computation Time', fontsize=40, fontweight='bold',
               labelpad=10, fontname='Times New Roman', color='green')

ax3.tick_params(axis='y', direction='out', width=3, length=10)
ax3.tick_params(axis='x', direction='out', width=3, length=10)

float_list3 = list(np.arange(0, 1.4, 0.2))
tick_labels3 = [f"{value:.1f}s" for value in float_list3]
ax3.set_yticks(float_list3)
ax3.set_yticklabels(tick_labels3, rotation=0, fontsize=30, color='blue',
                    fontname='Times New Roman', ha='right')

ax3.set_xticks([1]+list(range(10,60,10)))
ax3.set_xticklabels([1]+list(range(10,60,10)), rotation=0, fontsize=40, color='blue',
                    fontname='Times New Roman', ha='right', fontweight='bold')

lines3, labels3 = ax3.get_legend_handles_labels()
legend3 = ax3.legend(lines3, labels3, loc='upper left',
                    frameon=True, shadow=True, markerscale=3,
                    fancybox=True, prop={'family': 'Times New Roman', 'size': 30})
legend3.get_frame().set_color('#E6E6FA')
legend3.get_frame().set_alpha(0.8)
ax3.grid(True, which='both', linestyle='-', linewidth=2, alpha=0.3)

# ********************************************************************************************
sns.lineplot(x=dim4, y=Main_times[0:200], ax=ax4, label='Our Result', markerfacecolor='#FF6347',
             linestyle='-', marker='v', color='#20B2AA',
             markersize=10, linewidth=6)

ax4.set_xlabel(f'Matrix Dimension', fontsize=40, fontweight='bold',
               labelpad=0, fontname='Times New Roman', color='green')
# ax4.set_ylabel(f'Computation Time', fontsize=40, fontweight='bold',
#                labelpad=10, fontname='Times New Roman', color='green')

ax4.tick_params(axis='y', direction='out', width=3, length=10)
ax4.tick_params(axis='x', direction='out', width=3, length=10)

float_list4 = list(np.arange(0, 0.42, 0.06))
tick_labels4 = [f"{value:.2f}s" for value in float_list4]
ax4.set_yticks(float_list4)
ax4.set_yticklabels(tick_labels4, rotation=0, fontsize=30, color='blue',
                    fontname='Times New Roman', ha='right')

ax4.set_xticks([1]+list(range(20,220,20)))
ax4.set_xticklabels([1]+list(range(20,220,20)), rotation=0, fontsize=40, color='blue',
                    fontname='Times New Roman', ha='right', fontweight='bold')

lines4, labels4 = ax4.get_legend_handles_labels()
legend4 = ax4.legend(lines4, labels4, loc='upper left',
                    frameon=True, shadow=True, markerscale=3,
                    fancybox=True, prop={'family': 'Times New Roman', 'size': 30})
legend4.get_frame().set_color('#E6E6FA')
legend4.get_frame().set_alpha(0.8)
ax4.grid(True, which='both', linestyle='-', linewidth=2, alpha=0.3) 
# ********************************************************************************************
fig.suptitle(f'Computation Time Using Different Methods',
          fontname='Times New Roman', fontsize=60, fontweight='bold', color='green')

# plt.savefig('Fig4.pdf', format='pdf', edgecolor='black', dpi=800)
plt.show()
```
