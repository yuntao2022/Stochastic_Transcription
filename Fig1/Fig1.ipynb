## **Complete Sample Paths**

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.load('sample_paths_SSA.npy', allow_pickle=True).item()

time_point = data['time']
mRNA_data = data['mRNA']

print("Shape of time data:", time_point.shape)
print("Shape of mRNA number data:", mRNA_data.shape)
```

```python
ode_data = np.load('sample_paths_ode.npy')
ode_data = ode_data.tolist()
```

```python
plt.figure(figsize=(18, 11), dpi=100)
plt.style.use('default')
# plt.style.use('ggplot')
# The following list contains 20 different colors
colors = [
    '#A52A2A','#653700','#0343DF','#ADD8E6','#7BC8F6','#9ACD32',
    '#FFA500','#800080','#FFC0CB','#D2B48C','#006400','#8B4513','#00FFFF',
    '#FF6347','#BA55D3','#A9A9A9','#4682B4','#3CB371','#FFD700','#8B0000'
]

# Specify the number of sample paths to plot. 
# Num should a nonnegative integer smaller than 20. Note that (Num+1) sample paths will be plotted in total.
Num=9


for i in range(Num):
    plt.plot(time_point, mRNA_data[i], color=colors[i], linestyle="-",
             linewidth=.5)
plt.plot(time_point, mRNA_data[-1], color="#C79FEF", linestyle="-",
         linewidth=.5, label='Sample Paths of Stochastic Description')
plt.plot(time_point,ode_data,linestyle="--",label='Deterministic Description',linewidth=6,c='#13EAC9')
# plt.plot(time_point,ode_data,linestyle="--",label='Deterministic Description',linewidth=6,c='darkgreen')

plt.title("Sample Paths from Stochastic Simulation",
          fontname='Times New Roman',
          # fontname='DejaVu Sans',
          fontsize=40,fontweight='bold',color='green')
plt.xlabel('Time', 
           fontname='Times New Roman',
           # fontname='DejaVu Sans',
           fontsize=40,fontweight='bold',color='green')
plt.ylabel('Number of mRNA Molecules', 
           fontname='Times New Roman',
           # fontname='DejaVu Sans',
           fontsize=40,fontweight='bold',color='green')
plt.xticks(
    ticks=list(range(0,41,10)),
               fontsize=40,
               color='blue',
               fontname='Times New Roman',
               # fontname='DejaVu Sans',
               ha='right')
plt.yticks(
    ticks=np.arange(0, 121, 20),
               fontsize=40,
               color='blue',
               rotation=90,
               fontname='Times New Roman',
               # fontname='DejaVu Sans',
               ha='right')
legend=plt.legend(loc='upper right', frameon=True, shadow=True, markerscale=1,
           fancybox=True,prop={'family': 'Times New Roman','size': 30})
legend.get_frame().set_color('#E6E6FA')
legend.get_frame().set_alpha(0.8)
# plt.savefig('Fig1.pdf', bbox_inches='tight',format='pdf',dpi=800)
plt.grid(True, which='both', linestyle='-', linewidth=2, alpha=0.2)
# The generated plot can be considerably large. Avoid displaying it directly in this Notebook if possible.
plt.show()
# plt.close()
```
